```python
import numpy as np
import matplotlib.pyplot as plt
def contractplot(control_group,contract_group,labelc,labelg,name):
    contro = np.load(control_group)
    contra = np.load(contract_group)
    fig, axes = plt.subplots(1, 3, figsize=(14,5))
    axes[0].plot(contro['lr'], label=labelc, color='blue')
    axes[0].plot(contra['lr'], label=labelg, color='red', linestyle='--')
    axes[0].set_title('LR Schedule Comparison'); axes[0].legend(); axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Learning Rate')
    
    axes[1].plot(contro['test_acc'], label=labelc, color='blue')
    axes[1].plot(contra['test_acc'], label=labelg, color='red', linestyle='--')
    axes[1].set_title('Test Accuracy Comparison'); axes[1].legend(); axes[1].set_xlabel('Epoch')

    axes[2].plot(contro['loss'], label=labelc, color='blue')
    axes[2].plot(contra['loss'], label=labelg, color='red', linestyle='--')
    axes[2].set_title('Loss Curve Comparison'); axes[2].legend(); axes[2].set_xlabel('Epoch')
    axes[2].set_ylabel('Loss')
    axes[2].set_yscale('log')
    plt.savefig(name+'.png')
    plt.show()
contractplot('history_Sin-SGDR_cifar10_lrtuned_T1.npz','history_Standard-SGDR_cifar10_lrtuned_T1.npz','Sin-SGDR','Standard-SGDR','cifar-10-sin-stand')
contractplot('history_Sin-SGDR_cifar100_lrtuned_T1.npz','history_Standard-SGDR_cifar100_lrtuned_T1.npz','Sin-SGDR','Standard-SGDR','cifar-100-T1-sin-stand')
contractplot('history_Sin-SGDR_cifar100_lrtuned_T2.npz','history_Standard-SGDR_cifar100_lrtuned_T2.npz','Sin-SGDR','Standard-SGDR','cifar-100-T2-sin-stand')
contractplot('history_Maxlr-SGDR_cifar10_lr_range.npz','history_Minlr-SGDR_cifar10_lr_range.npz','Max-lr','Min-lr','cifar-10-lr-range')
contractplot('history_Maxlr-SGDR_cifar100_lr_range_T1.npz','history_Minlr-SGDR_cifar100_lr_range_T1.npz','Max-lr','Min-lr','cifar-100-T1-lr-range')
contractplot('history_Maxlr-SGDR_cifar100_lr_range_T2.npz','history_Minlr-SGDR_cifar100_lr_range_T2.npz','Max-lr','Min-lr','cifar-100-T2-lr-range')
contractplot('history_Sin-SGDR_cifar10_lrtuned_T1.npz','history_lineardown_cifar10_t1_cifar10_lrtuned_T1.npz','Sin-SGDR','Lineardecay-SGDR','cifar-10-sin-lindown')
contractplot('history_Sin-SGDR_cifar100_lrtuned_T1.npz','history_lineardown_cifar100_t1_cifar10_lrtuned_T1.npz','Sin-SGDR','Lineardecay-SGDR','cifar-100-T1-sin-lindown')
contractplot('history_Sin-SGDR_cifar100_lrtuned_T2.npz','history_lineardown_cifar100_t2_cifar10_lrtuned_T1.npz','Sin-SGDR','Lineardecay-SGDR','cifar-100-T2-sin-lindown')
```
